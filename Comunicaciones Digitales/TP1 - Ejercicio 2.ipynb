## Item a
Son antipodales, porque $s_1(t) = - s_2(t)$.

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
from scipy import signal
```

## Item b

```python
s1 = np.array([0.25, 0.5, 0.5, 1, 1, 1, 1, 0.5, 0.5, 0.25])
```

```python
s2 = -s1
```

## Item c

```python
s1_2 = s1 - s2
```

```python
h_t = np.flip(s1_2)
print(f'h(t): {h_t}')
```

## Item d
$$\gamma = 0$$

```python
a1_t = signal.lfilter(h_t, [1], s1)
a2_t = signal.lfilter(h_t, [1], s2)
```

```python
a1 = a1_t[-1]
a2 = a2_t[-1]
print(f'a1 = {a1}')
print(f'a2 = {a2}')
```
