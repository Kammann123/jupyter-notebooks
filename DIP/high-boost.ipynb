```python
import cv2
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('cactus.png', cv2.IMREAD_GRAYSCALE)
```

```python
image = image[:400,:400]
```

```python
plt.imshow(image, cmap='gray')
```

```python
kernel = np.ones((9, 9))
kernel /= kernel.sum()
image = cv2.filter2D(src=image, ddepth=-1, kernel=kernel)
plt.imshow(image, cmap='gray')
```

```python
A = 1.8
kernel = -np.ones((3, 3))
kernel[1, 1] = A + 8
```

```python
filtered = cv2.filter2D(src=image, ddepth=-1, kernel=kernel)
```

```python
plt.imshow(filtered, cmap='gray')
```
