```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
# Se registran las resoluciones como (ancho,alto,relacionAncho,relacionAlto)
resoluciones = [(320,240,4,3),     # 240p 
                (640,480,4,3),     # 480p
                (1280,720,16,9),   # 720p
                (1920,1080,16,9),  # 1080p
                (3840,2160,16,9),  # 4k
               ]

# Creo arreglo con las diagonales a utilizar de monitores
diagonales = np.linspace(0.3, 3, 100)

# Creo el arreglo para guardar resultados para el gr√°fico
distancias = np.zeros((len(resoluciones), len(diagonales)))

for i in range(len(resoluciones)):
    # Calculo la pendiente
    pw,ph,rw,rh = resoluciones[i]
    pendiente = (1 / (np.tan(np.deg2rad(1/120)) * 2)) * (rw / pw) / np.sqrt(rw**2 + rh**2)
    
    # Calculo la distancia
    distancias[i,:] = pendiente * diagonales
```

```python
# Grafico los resultados
plt.figure(figsize=(18, 9))

for i in range(len(resoluciones)):
    plt.plot(diagonales, distancias[i,:], label=f'{resoluciones[i][0]}x{resoluciones[i][1]}')

plt.xlabel('Diagonal [m]', fontsize=16)
plt.ylabel('Distancia [m]', fontsize=16)
plt.legend(fontsize=16)
plt.grid()
plt.show()
```
