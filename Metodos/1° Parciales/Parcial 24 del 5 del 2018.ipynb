```python
# Ejercicio 1
from numpy import *
from scipy import linalg

A = array(
    [
        [1, 1],
        [0, 1],
        [0, -1],
        [-1, -1]
    ]
)

b = array(
    [
        [16],
        [8],
        [-8],
        [-32]
    ]
)

Q = array(
    [
        [1 / sqrt(2), 0],
        [0, 1 / sqrt(2)],
        [0, -1 / sqrt(2)],
        [-1 / sqrt(2), 0]
    ]
)

R = array(
    [
        [sqrt(2), sqrt(2)],
        [0, sqrt(2)]
    ]
)

linalg.solve(R, Q.T @ b)
```

```python
# Ejercicio 2
def punto_fijo(x0: float, g: callable, iter_count: int) -> float:
    """ Implementación del método numérico para resolver ecuaciones no lineales
        por punto fijo.
        :parametro x0: Condicion inicial
        :parametro g: Función de la ecuación a resolver g(x) = x
        :parametro iter_count: Contador de iteraciones
    """
    x = [x0]
    for i in range(1, 1 + iter_count):
        x.append(g(x[i - 1]))
    return x

x0 = 0.653
g = lambda x: exp(- (x**2))
iter_count = 10
results = punto_fijo(x0, g, iter_count)

print("Resultado: {}".format([round(value, 5) for value in results]))
```
