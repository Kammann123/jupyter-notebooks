```python
from matplotlib import pyplot
from numpy import arange, sin, pi

time = arange(0, 2 * pi, 1 / 1000)
y = [sin(t) for t in time]

def muestreo_natural(tau, time, y):
    values = []
    for i, t in enumerate(time):
        if t < tau:
            values.append(y[i])
        else:
            values.append(0)
    return values

def muestreo_instantaneo(tau, time, y):
    values = []
    for i, t in enumerate(time):
        if t == tau:
            values.append(y[i])
        else:
            values.append(0)
    return values

pyplot.plot(time, y)
pyplot.plot(time, muestreo_instantaneo(0.9, time, y))
pyplot.plot(time, muestreo_natural(0.6, time, y))
pyplot.grid()
pyplot.show()
```
