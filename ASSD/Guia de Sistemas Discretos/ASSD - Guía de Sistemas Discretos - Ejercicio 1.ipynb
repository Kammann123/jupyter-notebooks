1) Analizar para los siguientes filtros invariancia en el tiempo, causalidad y linealidad. Se entregan los siguientes items:

* (d) $R[x(n \cdot T)](n) = 5 \cdot n \cdot T \cdot x^{2}(n \cdot T)$
* (e) $R[x(n \cdot T)](n) = 3 \cdot x(n \cdot T + 3 \cdot T)$
* (i) $R[x(n \cdot T)](n) = x(n \cdot T + T) \cdot e^{-n \cdot T}$

A modo de aclaración, acá se decide utilizar la notación $R[x(n \cdot T)](n)$, donde se aplica una entrada $x(n \cdot T)$ al sistema, y dicha salida es evaluada en el instante $n$ del tiempo discreto.

# Ejercicio 1.d
El sistema no es invariante en el tiempo, es causal y no es lineal.

##### Invariancia en el tiempo
\begin{equation}
    R[x(n \cdot T - \alpha \cdot T)](n) = 5 \cdot n \cdot T \cdot x^{2}(n \cdot T - \alpha \cdot T) \\
    R[x(n \cdot T)](n - \alpha) = 5 \cdot (n - \alpha) \cdot T \cdot x^{2}(n \cdot T - \alpha \cdot T)
\end{equation}

Entonces como, $R[x(n \cdot T - \alpha \cdot T)](n) \neq R[x(n \cdot T)](n - \alpha)$, se dice que no es de tiempo invariante, lo cual resulta evidente dado que los coeficientes del sistema dependen de la variable temporal.

##### Causalidad
Sean dos entradas $x_1$ y $x_2$, donde $x_1(n) = x_2(n)$ $\forall n < k, k \in Z$. Entonces, evaluando en el sistema para ver cómo son las salidas:

\begin{equation}
    R[x_1(n \cdot T)](n) = 5 \cdot n \cdot T \cdot x_1^{2}(n \cdot T) \\
    R[x_2(n \cdot T)](n) = 5 \cdot n \cdot T \cdot x_2^{2}(n \cdot T)
\end{equation}

Entonces, se puede decir que $R[x_1(n \cdot T)](n) = R[x_2(n \cdot T)](n)$ $\forall n < k, k \in Z$, puesto que siempre para un instante $n$, el sistema analiza ese instante, con lo cual siempre que se analice hasta donde las señales son iguales, las salidas son iguales. En conclusión, el sistema es causal.

##### Linealidad

\begin{align}
    R[\alpha_1 \cdot x_1(n \cdot T) + \alpha_2 \cdot x_2(n \cdot T)](n) &= 5 \cdot n \cdot T \cdot [\alpha_1 \cdot x_1 + \alpha_2 \cdot x_2]^{2}(n \cdot T) \\
    &= 5 \cdot n \cdot T \cdot [(\alpha_1 \cdot x_1)^{2} + (\alpha_2 \cdot x_2)^{2} + 2 \cdot \alpha_1 \cdot \alpha_2 \cdot x_1 \cdot x_2](n \cdot T) \\
    &= \alpha_1^{2} \cdot R[x_1(n \cdot T)](n \cdot T) + \alpha_2^{2} \cdot R[x_2(n \cdot T)](n \cdot T) + 5 \cdot n \cdot T \cdot 2 \cdot \alpha_1 \cdot \alpha_2 \cdot x_1(n \cdot T) \cdot x_2(n \cdot T) \\
    &\neq \alpha_1 \cdot R[x_1(n \cdot T)](n \cdot T) + \alpha_2 \cdot R[x_2(n \cdot T)](n \cdot T)
\end{align}

Entonces, como ante una dada combinación lineal de entrada con coeficientes $\alpha_1, \alpha_2 \in R$ luega la salida no corresponde a la combinación lineal de la respuesta a las entradas individualmente, luego este sistema no es lineal.


# Ejercicio 1.e
El sistema es de tiempo invariante, no es causal y es lineal.

##### Invariancia en el tiempo

\begin{equation}
    R[x(n \cdot T - \alpha \cdot T](n) = 3 \cdot x(n \cdot T + 3 \cdot T - \alpha \cdot T) \\
    R[x(n \cdot T)](n - \alpha) = 3 \cdot x(n \cdot T - \alpha \cdot T + 3 \cdot T)
\end{equation}

Entonces como, $R[x(n \cdot T - \alpha \cdot T)](n) = R[x(n \cdot T)](n - \alpha)$, se dice que es de tiempo invariante, lo cual resulta evidante dado que los coeficientes del sistema son constantes y no varian con la variable temporal.

##### Causalidad
Sean dos entradas $x_1$ y $x_2$, donde $x_1(n) = x_2(n)$ $\forall n < k, k \in Z$. Entonces, evaluando en el sistema para ver cómo son las salidas:

\begin{equation}
    R[x_1(n \cdot T)](n) = 3 \cdot x_1(n \cdot T + 3 \cdot T) \\
    R[x_2(n \cdot T)](n) = 3 \cdot x_2(n \cdot T + 3 \cdot T)
\end{equation}

Entonces, se puede decir que $R[x_1(n \cdot T)](k) \neq R[x_2(n \cdot T)](k)$ dado que el sistema evalúa las entradas en el instante $n = k + 3$, dónde ya no son iguales. Ergo, el sistema no es causal, lo cual resulta evidente porque el sistema observa sucesos futuros o posteriores al instante $n$.


##### Linealidad
\begin{equation}
    R[\alpha_1 \cdot x_1(n \cdot T) + \alpha_2 \cdot x_2(n \cdot T)](n) = 3 \cdot [\alpha_1 \cdot x_1 + \alpha_2 \cdot x_2](n \cdot T + 3 \cdot T) = \alpha_1 \cdot R[x_1(n \cdot T)](n) + \alpha_2 \cdot R[x_2(n \cdot T)](n)
\end{equation}

Entonces, como ante una dada combinación lineal de entrada con coeficientes $\alpha_1, \alpha_2 \in R$ luega la salida es combinación lineal de la respuesta a las entradas individualmente, luego este sistema es lineal.

# Ejercicio 1.i
El sistema es lineal, no causal y no es tiempo invariante.

##### Invariancia en el tiempo
\begin{equation}
    R[x(n \cdot T - \alpha \cdot T)](n) = x(n \cdot T + T - \alpha \cdot T) \cdot e^{-n \cdot T} \\
    R[x(n \cdot T)](n - \alpha) = x(n \cdot T - \alpha \cdot T + T) \cdot e^{-n \cdot T + \alpha \cdot T}
\end{equation}

Entonces como, $R[x(n \cdot T - \alpha \cdot T)](n) \neq R[x(n \cdot T)](n - \alpha)$, se dice que no es de tiempo invariante, lo cual resulta evidante dado que los coeficientes son función de la variable temporal.

##### Causalidad
Sean dos entradas $x_1$ y $x_2$, donde $x_1(n) = x_2(n)$ $\forall n < k, k \in Z$. Entonces, evaluando en el sistema para ver cómo son las salidas:

\begin{equation}
    R[x_1(n \cdot T)](n) = x_1(n \cdot T + T) \cdot e^{-n \cdot T} \\
    R[x_2(n \cdot T)](n) = x_2(n \cdot T + T) \cdot e^{-n \cdot T}
\end{equation}

Entonces, se puede decir que $R[x_1(n \cdot T)](k) \neq R[x_2(n \cdot T)](k)$ dado que el sistema evalúa las entradas en el instante $n = k + 1$, dónde ya no son iguales. Ergo, el sistema no es causal, lo cual resulta evidente porque el sistema observa sucesos futuros o posteriores.

##### Linealidad
\begin{equation}
    R[\alpha_1 \cdot x_1(n \cdot T) + \alpha_2 \cdot x_2(n \cdot T)](n) = [\alpha_1 \cdot x_1 + \alpha_2 \cdot x_2](n \cdot T +  T) \cdot e^{-n \cdot T} = \alpha_1 \cdot R[x_1(n \cdot T)](n) + \alpha_2 \cdot R[x_2(n \cdot T)](n)
\end{equation}

Entonces, como ante una dada combinación lineal de entrada con coeficientes $\alpha_1, \alpha_2 \in R$ luega la salida es combinación lineal de la respuesta a las entradas individualmente, luego este sistema es lineal.
