# Resolución
En primer lugar, desde un punto de vista temporal, definir la señal de salida requiere de definir una señal auxiliar o intermedia en el punto medio del sistema, ya que esa señal es realimentada. A partir de aquí se consideran $x, y$ señales de tiempo discreto donde $x(n) = x_T(n \cdot T)$ y $y(n) = y_T(n \cdot T)$ para caracterizar el muestreo en el tiempo de las señales continuas.

\begin{align}
    & y(n) = p(n) + p(n-1) \\
    & p(n) = x(n) + p(n-1) - \frac{1}{2} \cdot p(n-2)
\end{align}

En primer lugar, a partir de estas dos expresiones se podría considerar al sistema como la conexión en cascada de dos sistemas distintos, donde ambos son lineales, causales y de tiempo invariante. Por ende, se decide pasar al dominio de la transforamada Z para facilicar encontrar la ecuación a diferencias. Para ello, se asumen causalidad y sistema relajado para considerar condiciones iniciales nulas.

\begin{align}
    & Y(z) = P(z) + P(z) \cdot z^{-1} = P(z) \cdot (1 + z^{-1}) \\
    & P(z) = X(z) + P(z) \cdot (z^{-1} - \frac{1}{2} \cdot z^{-2}) \Rightarrow P(z) = \frac{X(z)}{1 - z^{-1} + \frac{1}{2} \cdot z^{-2}} 
\end{align}

\begin{equation}
    Y(z) = X(z) \cdot \frac{1 + z^{-1}}{1 - z^{-1} + \frac{1}{2} \cdot z^{-2}} \Rightarrow 
    Y(z) \cdot (1 - z^{-1} + \frac{1}{2} \cdot z^{-2}) = X(z) \cdot (1 + z^{-1})
\end{equation}

Luego, antitransformando, se encuentra la ecuación a diferencias del sistema:

\begin{equation}
    y(n) - y(n-1) + 0.5 \cdot y(n-2) = x(n) + x(n-1)
\end{equation}
