```python
from scipy import signal

fs = 48000
edges = [0, 2000, 2200, 3000, 3200, 24000]
w = [0, 1, 0]
weight = [0.1, 0.1, 0.1]
n = 350

h = signal.remez(n, edges, w, weight=weight, Hz=fs)
```

```python
from matplotlib import pyplot
import numpy as np

w, H= signal.freqz(h, [1], worN=4096)

fig, ax = pyplot.subplots(1, 1)

fig.set_figwidth(16)
fig.set_figheight(7)

ax.plot(w * fs / (2 * np.pi), abs(H))
ax.minorticks_on()
ax.grid(which="both")
ax.set_xlabel("Frecuencia [Hz]")

pyplot.show()
```
