```python
from scipy import signal
import numpy as np

s = signal.lti([1, 0], [1, 400, 20000])
z = signal.dlti([1e-3 * 1,1e-3 * -0.991, 0], [1, -1.653, 0.6702], dt=1e-3)
```

```python
from matplotlib import pyplot

s_w, s_mag, s_phase = s.bode()
z_w, z_mag, z_phase = z.bode()

pyplot.semilogx(s_w / (2 * np.pi), s_mag)
pyplot.semilogx(z_w / (2 * np.pi), z_mag)

pyplot.show()
```

```python
from scipy import signal
import numpy as np

T = 2
a = 0.05
b = 0.1
s = signal.ZerosPolesGain(
    [(a + b) / 2],
    [a, b],
    2
)
alfa = np.exp(a * T)
beta = np.exp(b * T)
z = signal.ZerosPolesGain(
    [0, (alfa + beta) / 2],
    [alfa, beta],
    8,
    dt=2
)
alfa = np.tan(a)
beta = np.tan(b)
gamma = np.tan((a + b) / 2)
z = signal.ZerosPolesGain(
    [-1, (1 + gamma) / (1 - gamma)],
    [(1 + alfa) / (1 - alfa), (1 + beta) / (1 - beta)],
    2 * (1 - gamma) / (1 - alfa) / (1 - beta),
    dt=2
)
```

```python
from matplotlib import pyplot

s_w, s_mag, s_phase = s.bode()
z_w, z_mag, z_phase = z.bode()
pyplot.semilogx(s_w / (2 * np.pi), s_mag)
pyplot.semilogx(z_w / (2 * np.pi), z_mag)

pyplot.show()
```
