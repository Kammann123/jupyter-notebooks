# Sistema I

### Respuesta impulsiva
Dado que el sistema descripto por la respuesta impulsiva $h(n)$ es LTI, luego la respuesta a una excitación $x(n)$ está definida por la convolución, de esta forma se procede a obtener la respuesta del sistema $y(n)$.

\begin{align*}
    y(n) &= r(-n) = [g(-k) \circledast h(k)] \rvert_{k = -n} = g(n) \circledast h(-n) \\
    &= x(n) \circledast h(n) \circledast h(-n) \\
    &= x(n) \circledast h_I(n)
\end{align*}

Por lo tanto, la respuesta impulsiva por definición es,

\begin{equation}
    h_I(n) = h(n) \circledast h(-n)
\end{equation}

Es de interés remarcar que la respuesta impulsiva es una función par, dado que $h_I(n) = h_I(-n)$ y como tal, no es causal. Por ende, el sistema resultante no es causal.

### Respuesta en frecuencia
En primer lugar, se desea determinar si el resultado es un sistema estable. Para ello se busca demostrar que existe la transformada Z del sistema obtenido, y en consecuencia observar que la región de convergencia del mismo contiene el círculo de radio unitario. Por ende, no sólo converge y es estable, sino que además sobre tal espacio se encuentra la respuesta en frecuencia del sistema.

Como $h(n)$ describe un sistema estable y causal, luego la transformada Z del mismo converge en el exterior de un circulo de radio menor que uno, pues incluye el circulo unitario.

\begin{align*}
    H(z) &= \sum_{n \in Z} h(n) \cdot z^{-n} , \forall z : |z| > r_1, 0 < r_1 < 1
\end{align*}

\begin{align*}
    Z[h(-n)](z) &= \sum_{n \in Z} h(-n) \cdot z^{-n} \\
    &= \sum_{n = -\infty}^{0} h(-n) \cdot z^{-n} \\
    &= \sum_{n = 0}^{+\infty} h(n) \cdot z^{n} \\
    &= \sum_{n = 0}^{+\infty} h(n) \cdot (\frac{1}{z})^{-n} \\
    &= H(\frac{1}{z}), \forall z : |z| < \frac{1}{r_1} \\
\end{align*}

Entonces, se puede observar que ambas $H(z)$ y $H(\frac{1}{z})$ poseen diferentes regiones de convergencia y eso provoca que debamos estudiar dónde converge $h_I(n)$. Esto último se puede observar sencillamente aplicando la definición de T.Z sobre la expresión encontrada para tal respuesta impulsiva.

\begin{align*}
    Z[h_I(n)] &= \sum_{n \in Z} h_I(n) \cdot z^{-n} = \sum_{n \in Z} \sum_{k \in Z} h(k) \cdot h(k - n) \cdot z^{-n} \\
    &= \sum_{n \in Z} \sum_{k \in Z} h(k) \cdot h(k - n) \cdot z^{k-n} \cdot z^{-k} \\
    &= \sum_{k \in Z} h(k) \cdot z^{-k} \sum_{n \in Z}  h(k - n) \cdot z^{k-n} \\
    &= \sum_{k \in Z} h(k) \cdot z^{-k} \sum_{i \in Z}  h(i) \cdot z^{i} \\
    &= H(z) \cdot H(\frac{1}{z}) = H_I(z)
\end{align*}

Esta última deducción pudo hacerse garantizando que cada serie de Laurent converja, para lo cual utilizando los resultados previos se encontró que la región de convergencia es el resultado de la intersección de ambas regiones dando así $r_1 < |z| < \frac{1}{r_1}$. Tal y como se puede observar en la siguiente imagen.

<img src="region.png">

Finalmente, la respuesta en frecuencia,

\begin{align*}
    &H_I(\omega) = H(z) \cdot H(\frac{1}{z}) \bigg \rvert_{z = e^{j \cdot \omega}} = |H(\omega)|^{2} \Rightarrow \\
    &|H_I(\omega)| = |H(\omega)|^{2} \\
    &Arg(H_I(\omega)) = 0 \\
\end{align*}

### Conclusión
El sistema obtenido se denomina un filtro de fase nula, el mismo permite eliminar las distorsiones provocadas sobre señales poliarmónicas por las discontinuidades de fase que se pueden producir entre los diferentes armónicos por la diferencias de retardo en tiempo o en fase al atravesar al sistema. La característica obtenida no obstante requiere que el sistema sea no causal, con lo cual estos sistemas deben ser utilizados en tiempo no real sobre muestras o secuencias finitas, así pueden utilizarse filtros con aproximaciones muy selectivas cuyos problemas de fase son compensados.

# Sistema II

### Respuesta impulsiva
Dado que la $h(n)$ es respuesta impulsiva de sistema lineal, tiempo invariante y causal, al igual que antes se puede aplicar la convolución para conocer la respuesta ante una dada excitación. Operando sobre el sistema propuesto, se obtiene que,

\begin{align*}
    y(n) &= g(n) + r(-n) = x(n) \circledast h(n) + x(n) \circledast h(-n) \\
    &= x(n) \circledast [h(n) + h(-n)] \\
    &= x(n) \cdot h_{II}(n)
\end{align*}

Por lo tanto,
\begin{equation}
    h_{II}(n) = h(n) + h(-n)
\end{equation}

Es de interés remarcar que, al igual que antes, se trata de una respuesta impulsiva no causal. Es una función par.

### Respuesta en frecuencia
A modo de simplicación, se aplican los mismos criterios de convergencia que para el sistema I, con lo cual aplicando directamente la transformada Z sobre el sistema, el resultado nos permite conocer

\begin{align*}
    &H_{II}(z) = H(z) + H(\frac{1}{z}) \Rightarrow \\
    &H_{II}(\omega) = H(\omega) + \overline{H(\omega)} = 2 \cdot Re[H(\omega)] \\
    &|H_{II}(\omega)| = 2 \cdot |Re[H(\omega)]| \\
    &Arg(H_{II}(\omega)) = Arg(Re[H(\omega)]) = k \cdot \pi, k \in Z \\
\end{align*}

### Conclusión
El sistema II a diferencia del sistema I, no presenta fase nula, sino que presenta una fase constante de a tramos. Esto implica que luego su retardo de grupo es nulo. Con lo cual, para un grupo de frecuencias en un entorno el retardo o desfase es el mismo y la distorsión se ve reducida.

# Diseño del filtro
Para realizar el diseño de Legendre primero defino el desarrollo necesario de la teoría de aproximaciones usando Legendre para el sistema analógico. Posteriormente defino el prewarping y la transformación bilineal para poder hacer las modificaciones a los requisitos y resultados del sistema analógico para pasar al dominio digital. Por último, obtengo las respuestas en frecuencia e impulsivas de los sistemas y las grafico.

```python
# Defino código propio para poder utilizar la aproximación de Legendre
# que de forma completa no se encuentra en Scipy.signal. Básicamente, se arma
# una respuesta en frecuencia utilizando el n-ésimo polinomio de Legendre que se obtenía
# de forma diferente para caso par o impar. Defino ambas formas y luego se itera para encontrar un orden
# que satisfaga la plantilla.
from scipy import signal
from scipy import special
from numpy import *

def epsilon(ap: float) -> float:
    """ Calculo el valor del factor epsilon que define la máxima
        atenuación en la banda pasante.
        :parametro ap: Máxima atenuación en banda de paso
    """
    return sqrt(10 ** (ap / 10) - 1)

def even_ln(n: int) -> poly:
    """ Calculo los polinomios integrados de Legendre para un orden par. 
        :parametro n: Orden de la función transferencia
    """
    if (n % 2) == 0:
        # El integrando corresponde a una serie de polinomios de legendre
        k = n // 2 - 1
        b0 = 1 / sqrt((k + 1) * (k + 2))
        poly = poly1d([b0 if (k % 2) == 0 else 0])

        for i in range(1, k + 1):
            if ((k % 2) == 1 and (i % 2) == 0) or ((k % 2) == 0 and (i % 2) == 1):
                continue
            bi = b0 * (2 * i + 1)
            new_poly = bi * special.legendre(i)
            poly = polyadd(poly, new_poly)

        poly = polymul(poly, poly)
        poly = polymul(poly, poly1d([1, 1]))

        # Computo la integral de la serie de polinomios de Legrende
        poly = polyint(poly)
        upper = poly1d([2, 0, -1])
        lower = poly1d([-1])

        # Evalúo la integral definida con Barrow
        return polysub(polyval(poly, upper), polyval(poly, lower))
    else:
        raise ValueError("¡Esta sólo funciona para ordenes pares!")

def odd_ln(n: int) -> poly:
    """ Calculo los polinomios integrados de Legendre para un orden impar.
        :parametro n: Orden de la función transferencia
    """
    if n % 2:
        # El integrando corresponde a una serie de polinomios de legendre
        k = (n - 1) // 2
        a0 = 1 / (sqrt(2) * (k + 1))
        poly = poly1d([a0])

        for i in range(1, k + 1):
            ai = a0 * (2 * i + 1)
            poli = special.legendre(i)
            new_poly = ai * poli
            poly = polyadd(poly, new_poly)

        poly = polymul(poly, poly)

        # Computo la integral de la serie de polinomios de Legrende
        poly = polyint(poly)
        upper = poly1d([2, 0, -1])
        lower = poly1d([-1])

        # Evalúo la integral definida con Barrow
        return polysub(polyval(poly, upper), polyval(poly, lower))
    else:
        raise ValueError("¡Esta sólo funciona para ordenes impares!")

def design_legendre(n: int, ap: float):
    """ Diseño un filtro pasabajos normalizado de Legendre.
        :parametro n: Orden de la función transferencia
        :parametro ap: Máxima atenuación en banda de paso
    """
    # Calculamos parámetros de interés
    eps = epsilon(ap)
    zero_limit = 1e-10
    
    # Calculo el Ln, y construyo el denominador de |H(w)|^2
    ln = odd_ln(n) if (n % 2) else even_ln(n)
    ln = (eps ** 2) * ln
    den = polyadd(poly1d([1]), ln)
    
    # Normalizando y pasando los polos al plano complejo s
    gain = 1
    poles = []
    for pole in 1j * den.roots:
        if pole.real < 0:
            _pole = complex(
                pole.real if abs(pole.real) > zero_limit else 0,
                pole.imag if abs(pole.imag) > zero_limit else 0,
            )
            gain /= abs(_pole)
            poles.append(_pole)
    
    # Devuelvo el resultado
    return signal.ZerosPolesGain([], poles, 1 / gain)

def design_legendre_by_normalized_template(ap: float, wa: float, aa: float, max_order = 20):
    """ Diseño un filtro pasabajos normalizado de Legendre con los datos de la plantilla,
        básicamente se itera hasta encontrar el resultado correcto en orden.
        :parametro ap: Máxima atenuación en banda de paso
        :parametro wa: Frecuencia de banda de atenuación normalizada
        :parametro aa: Mínima atenuación en banda de atenuación
        :parametro max_order: Fija el maximo orden de iteración
    """
    for i in range(1, max_order + 1):
        h = design_legendre(i, ap)
        eval_wa, eval_aa, eval_phase = h.bode(w = [wa])
        if eval_aa < -aa:
            return h
    return None

def design_legendre_by_template(wp: float, ap: float, wa: float, aa: float, max_order = 20):
    """ Diseño un filtro pasabajos desnormalizado de Legendre con los datos de la plantilla,
        básicamente se itera hasta encontrar el resultado correcto en orden.
        :parametro wp: Frecuencia de banda de paso
        :parametro ap: Máxima atenuación en banda de paso
        :parametro wa: Frecuencia de banda de atenuación normalizada
        :parametro aa: Mínima atenuación en banda de atenuación
        :parametro max_order: Fija el maximo orden de iteración
    """
    # Normalizando la plantilla
    wan = wa / wp
    h = design_legendre_by_normalized_template(ap, wan, aa, max_order)
    
    if h is not None:
        # Aplico transformación para desnormalizar el filtro
        poles = []
        gain = 1
        for pole in wp * h.poles:
            poles.append(pole)
            gain /= abs(pole)

        # Devuelvo resultado
        return signal.ZerosPolesGain([], poles, 1 / gain)
    else:
        return None
```

```python
# En esta parte voy a realizar lo calculos para obtener mi filtro correspondiente en el campo digital,
# para ello primero defino dos funciones a utilizar luego, que son el prewarping y la transformación bilineal
# que voy a aplicar sobre cada polo del campo analogico para ver su polo digital.
import math

def prewarping(omega: float, fs: float) -> float:
    """ Dada una frecuencia no normalizada en una red digital, se le
        aplica la inversa del efecto warping. El prewarping para compensar
        efectos secundarios de la transformación bilineal.
        :parametro omega: Frecuencias del filtro digital
        :parametro fs: Frecuencia de muestreo
    """
    return 2 * fs * tan(omega / (2 * fs))

def bilineal(pole: complex, fs: float) -> complex:
    """ Dado un polo en el campo analogico, con la transformación bilineal
        se obtiene el polo en el dominio digital.
        :parametro pole: Polo de la H(s)
        :parametro fs: Frecuencia de muestreo
    """
    return (2 * fs + pole) / (2 * fs - pole)

# Defino los datos de la consigna, frecuencias y atenuaciones de la plantilla no normalizada
# para el filtro, así como la frecuencia de muestreo del sistema digital.
fp = 2000
wp = 2 * math.pi * fp
fa = 3800
wa = 2 * math.pi * fa
ap = 0.2
aa = 30
fs = 40000

# Computo el filtro analogico para comparar los resultados
H_analog = design_legendre_by_template(wp, ap, wa, aa)

# Computo el filtro analogico con el prewarping y aplico la transformación bilineal
H_auxiliar = design_legendre_by_template(prewarping(wp, fs), ap, prewarping(wa, fs), aa)

gain = 1
poles = []
for pole in H_auxiliar.poles:
    _pole = bilineal(pole, fs)
    poles.append(_pole)
    gain /= (1 - _pole)
    
H_digital = signal.ZerosPolesGain([], poles, 1 / gain, dt = 1 / fs)
```

```python
from matplotlib import pyplot
from matplotlib import patches

# Computando respuesta en frecuencia
w = arange(2 * math.pi * 1 / fs, math.pi, math.pi / 100000)
w_a, m_a, p_a = H_analog.bode(w * fs)
w_d, m_d, p_d = H_digital.bode(w)

# Graficando...
fig, (ax1, ax2) = pyplot.subplots(2, 1)

ax1.semilogx(w_a / (2 * math.pi), m_a, label="analog")
ax1.semilogx(w_d / (2 * math.pi), m_d, label="digital")
ax1.set_xlabel("Frequency [Hz]", fontsize=14)
ax1.set_ylabel("Magnitude [dB]", fontsize=14)
ax1.grid()
ax1.legend()

ax2.semilogx(w_a / (2 * math.pi), p_a, label="analog")
ax2.semilogx(w_d / (2 * math.pi), p_d, label="digital")
ax2.set_xlabel("Frequency [Hz]", fontsize=14)
ax2.set_ylabel("Phase [°]", fontsize=14)
ax2.grid()
ax2.legend()

# Dibujo la plantilla
ax1.add_patch(
     patches.Rectangle(
        (1, -120),
        1999,
        119.8,
        facecolor = (255 / 255, 140 / 255, 115 / 255, 0.5),
        fill=True
     ) )

ax1.add_patch(
     patches.Rectangle(
        (3800, 0),
        16200,
        -30,
        facecolor = (255 / 255, 140 / 255, 115 / 255, 0.5),
        fill=True
     ) )

pyplot.show()
```

# Obtención del Sistema II con el filtro de Legendre
A partir de la $H(z)$ se obtiene la función transferencia del sistema II, $H_{II}(z)$ dado que se encontró la expresión que lo define en el análisis teórico realizado previamente.


\begin{align*}
    H_I(z) &= H(z) + H(\frac{1}{z}) \\

## Resultados
Nótese que a continuación al sistema causal y no causal se le ingresa la misma señal de entrada que por poseer armónicos en la banda pasante, no posee distorsiones de fase ni atenuaciones. Pero en el sistema no causal no posee retardo alguno en el tiempo respecto de la entrada.

```python
# Primero, defino mi propio generador de señales poliarmónicas, y para asegurarme que el contenido se limita a la banda pasante
# lo construyo como una serie finita de Fourier. Recordando que fp = 2kHz.
def generator(fo: float, n_count: int, envelope: callable, fp: float, size: int, fs: float) -> array:
    """ Generador de señales poliarmonicas utilizando la frecuencia fundamental y una cantidad
        finita de armónicos múltiplos en frecuencia. Para sea un poco más divertido, le paso una función
        que me define la envolvente espectral, es decir la amplitud de cada armónico.
        :parametro fo: Frecuencia fundamental
        :parametro n_count: Cantidad de armonicos, va a estar limitado si se pasa fo * n_count > fp
        :parametro envelope: Función que espero usar envelope(i) para definir la proporción de cada armónico.
        :parametro fp: Frecuencia de paso
        :parametro size: Cantidad de muestras en una ventana finita
        :parametro fs: Frecuencia de muestreo
    """
    x = []
    for n in range(0, size + 1):
        value = 0
        for i in range(1, (n_count if n_count * fo < fp else fp // fo) + 1):
            value += envelope(i) * sin(2 * math.pi * fo * i * (n / fs)) 
        x.append(value)
    return x

# Construyo una función
fo = 200
count = 10
periods = 1
size = int(fs * periods / fo)
n = arange(size + 1)
x = generator(fo, count, lambda index: ((-1)**(index) - 1) * (1 / index), fp, size, fs)

# Le aplico el filtro digital causal de Legendre
t, y = signal.dlsim(H_digital, x)

# Le aplico el filtro digital no causal de Legendre
z = convolve(x, h1)
z = z[int(len(z) / 2 - size / 2):int(len(z) / 2 + size / 2) + 1]

fig, (ax1, ax2, ax3) = pyplot.subplots(3, 1)

ax1.step(n, x, linestyle = "", marker = ".", label="x(n)")
ax1.legend()
ax1.grid()
ax1.set_xlabel("Sample", fontsize=14)

ax2.step(n, y, linestyle = "", marker = ".", label="y(n) con digital causal")
ax2.legend()
ax2.grid()
ax2.set_xlabel("Sample", fontsize=14)

ax3.plot(n, z, linestyle = "", marker = ".", label="y(n) con digital no causal")
ax3.legend()
ax3.grid()
ax3.set_xlabel("Sample", fontsize=14)

pyplot.show()
```
