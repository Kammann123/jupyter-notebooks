```python
from numpy import pi, arange, exp
from matplotlib import pyplot

def x_n(amp: float, n: int) -> float:
    if n == 0:
        return (2 * amp) / (3 * pi)
    else:
        return (6 * amp) / (pi * (9 - 4 * n ** 2))

def series(amp: float, time: list) -> float:
    values = []
    for t in time:
        value = 0
        for n in range(-50, 50, 1):
            value += x_n(amp, n) * exp(complex(0, (2 / 3) * n * t))
        values.append(value)
    return values

time = arange(-3 * pi, 3 * pi, 6 * pi / 10000)
values = series(10, time)

pyplot.plot(time, values)
pyplot.grid()
pyplot.show()
```

```python
armonics = arange(0, 20, 1)
frequencies = armonics / (3 * pi)
density = [abs(x_n(10, n)) for n in armonics]

pyplot.stem(frequencies, density, use_line_collection=True)
pyplot.grid()
pyplot.show()
```
