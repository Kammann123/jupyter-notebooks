```python
from numpy import pi, arange, exp
from matplotlib import pyplot

pyplot.rcParams["figure.figsize"] = (8,7)

def x_n(amp: float, n: int) -> float:
    if n == 0:
        return (2 * amp) / (3 * pi)
    else:
        return (6 * amp) / (pi * (9 - 4 * n ** 2))

def series(amp: float, time: list) -> float:
    values = []
    for t in time:
        value = 0
        for n in range(-15, 15, 1):
            value += x_n(amp, n) * exp(0 + 1j * 2 / 3 * n * t)
        values.append(value)
    return values

time = arange(0, 20 * pi, 20 * pi / 20000)
values = series(10, time)

pyplot.plot(time, values)
pyplot.xlabel("Tiempo [s]", fontsize=20)
pyplot.ylabel("Señal seno 3/2", fontsize=20)
pyplot.grid()
pyplot.savefig("seno")
pyplot.show()
```

```python
armonics = arange(0, 20, 1)
frequencies = armonics / (3 * pi)
density = [abs(x_n(10, n)) for n in armonics]

pyplot.stem(frequencies, density, 'r', markerfmt='ro', use_line_collection=True)
pyplot.xlabel("Frecuencia [Hz]", fontsize=20)
pyplot.ylabel("Magnitud de armónicos", fontsize=20)
pyplot.grid()
pyplot.savefig("seno_potencia")
pyplot.show()
```
