```python
import numpy as np

def factorial(n: int) -> int:
    result = 1
    for i in np.arange(1, n + 1, 1):
        result *= i
    return result

def poisson(c: float, k: int):
    return (c**k) * (np.exp(-c)) / factorial(k)

def success_probability(c: float, accepted: int):
    result = 1
    for i in range(0, accepted + 1, 1):
        result -= poisson(c, i)
    return result
```

```python
from matplotlib import pyplot

interval = np.arange(1, 8, 1)
results = [1 - success_probability(2, b) for b in interval]

pyplot.plot([0, 7], [0.9, 0.9], linestyle='dashed', color='b')
pyplot.plot(interval, results, label='Probabilidad de aceptar', linestyle='dotted', color='r')
pyplot.xlabel("Cantidad máxima a aceptar")
pyplot.legend()
pyplot.grid()
pyplot.show()
```
