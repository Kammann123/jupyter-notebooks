```python
def accept_function(n: int, p: float) -> float:
    """
    Probabilidad de aceptación de un lote de n muestras considerando que se tiene
    probabilidad p de que un articulo sea defectuoso, tomando como criterio de aceptacion
    que sean 0, 1 o 2 las defectuosas en dicha muestra.
    Experimentos aleatorios identicos e independientes, problema Binomial.
    :param n: Cantidad de muestreo
    :param p: Probabilidad de artículo defectuoso
    :return: Probabilidad de ser aceptado el lote
    """
    return (1-p)**n + n * p * (1-p)**(n-1) + ((n * (n-1)) / 2) * p**2 * (1-p)**(n-2)
```

```python
from matplotlib import pyplot
from numpy import arange

samples = 10
scenarios = arange(0, 1, 1 / 1000)
results = [accept_function(samples, p) for p in scenarios]

pyplot.plot(scenarios, results)
pyplot.ylabel("Probabilidad de aceptación")
pyplot.xlabel("Probabilidad de artículo defectuoso")
pyplot.grid()
pyplot.show()
```
