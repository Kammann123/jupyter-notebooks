```python
import numpy as np

def binomial_min_n(p: float, target: float):
    return np.log(1 - target) / np.log(1 - p)

def poisson_min_n(p: float, target: float):
    return - (np.log(1 - target)) / p
```

```python
from matplotlib import pyplot

target = 0.9

interval = np.arange(0.1, 0.005, -0.005)
binomial = [binomial_min_n(p, target) for p in interval]
poisson = [poisson_min_n(p, target) for p in interval]

pyplot.plot(interval, binomial, label='Binomial', color='r')
pyplot.plot(interval, poisson, label='Poisson', color='b')
pyplot.legend()
pyplot.grid()
pyplot.show()
```
